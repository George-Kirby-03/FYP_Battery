```python
class Customer:
    def __init__(self, name, discount = 0):
        self.name = name
        self.discount = discount
    
    def give_discount(self, amount):
        self.discount += amount

class VIPCustomer(Customer):
    def __init__(self, name, trial, discount = 0):
        super(VIPCustomer, self).__init__(name, discount)
        self.ewwn = "Pooop"
        self.trial = trial
    def give_vip_discount(self, amount):
        self.discount += amount * 2
    
```

```python
vip = VIPCustomer("Alice", 10)
vip.give_vip_discount(5)
print(vip.name)          # Output: Alice
print(vip.discount)      # Output: 10
print(vip.ewwn)         # Output: Pooop
norm = Customer("Bob")
norm.give_vip_discount(5)  # This will raise an AttributeError
```
