```python
import torch
import numpy as np
```

```python
def info(tensor):
    if isinstance(tensor, torch.Tensor):
        print(f"Tensor Info: \n Shape: {tensor.shape} \n Dtype: {tensor.dtype} \n Device: {tensor.device} \n")
    else:
        print("Not a tensor")   
```

```python
data = [[1, 2], [3, 4]]
x_data = torch.tensor(data)
x_ones = torch.ones_like(x_data)
x_rand = torch.rand_like(x_data, dtype=torch.float)
print(f"Ones Tensor: \n {x_ones} \n")
print(f"Ones_like tensor: \n {x_rand} \n")
```

```python
shape = (2,3,2)
rand_tens = torch.rand(shape)   
print(rand_tens)
info(rand_tens)
```
