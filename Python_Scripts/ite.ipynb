```python
import torch
from torch.utils.data import Dataset
from torch.utils.data import DataLoader
```

```python
data = {"Name": 3, "Age": 4, "City": 5, "Country": 6, "Occupation": 7}
```

```python
class test_dataset(Dataset):
    def __init__(self, data, transform=None, target_transform=None):
        self.data = data
        self.name = list(self.data.keys())
        self.label = list(self.data.values())
        self.transform = transform
        self.target_transform = target_transform

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        return self.name[idx], self.label[idx]
```

```python
train_dataloader = DataLoader(test_dataset(data), batch_size=2, shuffle=True)
```

```python
train_features, train_labels = next(iter(train_dataloader))
print(f"Name: {train_features} \nLabel: {train_labels}")
```
